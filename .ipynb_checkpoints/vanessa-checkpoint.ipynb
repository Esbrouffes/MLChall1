```python
from sklearn.preprocessing import StandardScaler
tempCPA = new_temperature.values
std_scale = StandardScaler().fit(tempCPA)
tempCPA_scaled = std_scale.transform(tempCPA)
#print(tempCPA_scaled)

from sklearn.decomposition import PCA
pca = PCA().fit(tempCPA_scaled[:,:12])
plt.plot(np.cumsum(pca.explained_variance_ratio_))
pca.explained_variance_ratio_.cumsum()
```

```python
temp = pca.transform(tempCPA_scaled[:,:12])

plt.scatter(temp[:,0], temp[:,1], c=new_temperature['Moyenne'])
plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
plt.xlim(-6, 7)
plt.ylim(-3, 3)

for i in range(temp.shape[0]):
    plt.text(temp[i,0], temp[i,1], new_villes[i][0])
```

```python
components = pca.components_
for i, (x, y) in enumerate(zip(components[0,:], components[1,:])):
    plt.plot([0, x], [0, y], color='k')
    plt.text(x, y, new_temperature.columns[i])

plt.plot([-0.6, 0.6], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-0.6, 0.6], color='grey', ls='--')

plt.xlim(-0.5, 0.5)
plt.ylim(-0.5, 0.5);
```

```python
# D'aprs le cercle des corrlations obtenu et le graphique de la position des individus, on peut qualifier 
# les villes selon leur temprature. C'est--dire que les villes se situant  droite du graphique vont avoir 
# des tempratures plus leves que celle  gauches. En effet toutes les variables ont une corrlation positive 
# avec l'axe 1. De plus les variables qui concernent les mois d'hivers ont une corrlation positive avec l'axe 2 
# tandis que les variables qui concernent les mois d't ont une corrlation ngative avec l'axe 2. Ainsi les 
# villes en haut du graphique auront un hiver doux mais un t froid et les villes en bas du graphique auront
# un hiver froid et un t chaud. 
# En suivant ce rsonnement, Moscou est ville o les tempratures sont basses et avec des hivers trs froids et 
# des ts chauds. Au contraire de Sville qui est une ville o il semble faire chaud tout au long de l'anne.
```
