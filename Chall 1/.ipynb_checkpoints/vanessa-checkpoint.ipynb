```python
new_temperature = temperature.drop(temperature.columns[[0,17]], axis='columns') #on enleve la colonne des villes
#print(new_temperature)
```

```python
from sklearn.preprocessing import StandardScaler
tempCPA = new_temperature.values
std_scale = StandardScaler().fit(tempCPA)
tempCPA_scaled = std_scale.transform(tempCPA)
#print(tempCPA_scaled)

from sklearn.decomposition import PCA
pca = PCA().fit(tempCPA_scaled[:,:12])
plt.plot(np.cumsum(pca.explained_variance_ratio_))
pca.explained_variance_ratio_.cumsum()
```

```python
# La première composante présente environ 86% de la variance tandis que la seconde en présente environ 12 %
# A elles deux, elle forme 98% de la variance. Ces deux composantes sont donc suffisentes pour projeter nos données
```

```python
temp = pca.transform(tempCPA_scaled[:,:12])

plt.scatter(temp[:,0], temp[:,1], c=new_temperature['Moyenne'])
plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
plt.xlim(-6, 7)
plt.ylim(-3, 3)

for i in range(temp.shape[0]):
    plt.text(temp[i,0], temp[i,1], new_villes[i][0])
```

```python
components = pca.components_
for i, (x, y) in enumerate(zip(components[0,:], components[1,:])):
    plt.plot([0, x], [0, y], color='k')
    plt.text(x, y, new_temperature.columns[i])

plt.plot([-0.6, 0.6], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-0.6, 0.6], color='grey', ls='--')

plt.xlim(-0.5, 0.5)
plt.ylim(-0.5, 0.5);
```

```python
# D'après le cercle des corrélations obtenu et le graphique de la position des individus, on peut qualifier 
# les villes selon leur température. C'est-à-dire que les villes se situant à droite du graphique vont avoir 
# des températures plus élevées que celle à gauches. En effet toutes les variables ont une corrélation positive 
# avec l'axe 1. De plus les variables qui concernent les mois d'hivers ont une corrélation positive avec l'axe 2 
# tandis que les variables qui concernent les mois d'été ont une corrélation négative avec l'axe 2. Ainsi les 
# villes en haut du graphique auront un hiver doux mais un été froid et les villes en bas du graphique auront
# un hiver froid et un été chaud. 
# En suivant ce résonnement, Moscou est ville où les températures sont basses et avec des hivers très froids et 
# des étés chauds. Au contraire de Séville qui est une ville où il semble faire chaud tout au long de l'année.
```
