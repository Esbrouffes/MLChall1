# Machine Learning Challenge 2 : Youtube Comments Toxicity 



L’objectif de ce défi est de comprendre et prédire la notion de toxicité dans une vidéo
Youtube. Il s’agit de déterminer dans quelle mesure l’espace des commentaires des médias
français est touché par les débats brutaux et agressifs. Nous posons comme objectif de
comprendre où et quand l’agressivité en ligne est retrouvée et quels facteurs la limite et la
favorisent. D’abord, en abordant la question sous la forme d’un problème de régression
visant à prédire le nombre d’insultes présentes dans les commentaires associés à une
vidéo. Ensuite, en construisant puis prédisant un indice original de toxicité à partir des
variables présentes dans la base.

## Présentation et analyse des données 

### Importation des bibliothèques et chargement des données

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import shuffle
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, accuracy_score, f1_score, precision_score, recall_score, mean_squared_error
from scipy.stats import linregress
import matplotlib as mpl
# import des données

pd.set_option('display.max_columns', None)
df = pd.read_csv('challenge_youtube_toxic.csv', encoding = 'latin1', engine = 'python', sep=';')
```

```python
df.head()
```

Il faut changer le type de certaines colonnes, et supprimer la colonne inutile : 

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

def convert_into_float(string):
    if type(string)==float:
        return string
    return float (string.replace(',','.'))

df['nbrMotMoyenne'] = df['nbrMotMoyenne'].apply(convert_into_float)
df['nbrMotInsulteMoyenne'] = df['nbrMotInsulteMoyenne'].apply(convert_into_float)
df['nbrMotAllongMoyenne'] = df['nbrMotAllongMoyenne'].apply(convert_into_float)
df['nbrMotMAJMoyenne'] = df['nbrMotMAJMoyenne'].apply(convert_into_float)
df['nbrExclMarkMoyenne'] = df['nbrExclMarkMoyenne'].apply(convert_into_float)
df['nbrQuestMarkMoyenne'] = df['nbrQuestMarkMoyenne'].apply(convert_into_float)
len(df.dropna()) # permet de déterminer la qualité de saisie de la donnée. Ici il n'y a donc pas de données manquantes
```

On va maintenant pouvoir calculer certaines statistiques sur les données afin de mieux les comprendre. 

```python
df.describe()
```

```python
videos_par_chaines=[len(df[df['channel_name'] == name].video_id.unique()) for name in df.channel_name.unique()]
print(' Nombre de chaînes différentes = ', len(df.channel_id.unique()), ", Nombre moyen de vidéos par chaîne : ", round(len(df)/len(df.channel_id.unique())) )
print("Ecart type du nombre de vidéos par chaînes : ", round( df.groupby('channel_name').video_id.count().std()))
```

On peut voir que le nombre de vidéos par chaîne est réparti sur une grande plage de données.
Trouvons maintenant la chaîne avec le plus d'abonnés. 

```python
# on trouve la chaîne YouTube avec le plus d'abonnés.

channel_max = ''
max_val = 0
for name in df.channel_name.unique() :
    if df[df['channel_name'] == name].subscriberCount.unique()[0] > max_val :
        max_val = df[df['channel_name'] == name].subscriberCount.unique()[0]
        channel_max = name
    else :
        max_val = max_val
# print(channel_max, max_val)
print('La chaîne qui a le maximum d\'abonnés est : ', channel_max, ' et a : ', max_val, 'abonnés.') 
```

La valeur réelle trouvée sur Youtube est de 900 000 abonnés environ. Cela nous indique que les données ont dû être récoltés dans les 2 dernières années (estimation), et ne sont donc pas obsolètes.

```python
# création d'une colonne indiquant un numéro unique pour chaque chaîne

from sklearn.preprocessing import LabelEncoder
le_channel_name = LabelEncoder()
df['ind'] = le_channel_name.fit_transform(df['channel_name'])
df
```

```python
# scatter plot des données pour essayer de voir si des clusters se forment selon des variables
slope, intercept, r_value, p_value, std_err = linregress(df['nbrMot'], df['nbrMotInsulte'])
fitLine = slope * df['nbrMot'] + intercept
fig, ax = plt.subplots(1, 1, figsize=(20,6))
ax.plot(df['nbrMot'], fitLine, c='b')
ax.scatter(df['nbrMot'], df['nbrMotInsulte'], c = list(df['ind']), cmap = 'winter')
```

```python
slope1, intercept1, r_value, p_value, std_err = linregress(df.groupby('channel_name')['subscriberCount'].mean(),
                                                         df.groupby('channel_name')['nbrMotInsulte'].sum())
fitLine = slope1 * df.groupby('channel_name')['subscriberCount'].mean() + intercept1
fig1, ax1 = plt.subplots(1, 1, figsize=(20,6))
ax1.plot(df.groupby('channel_name')['subscriberCount'].mean(), fitLine, c='b')
ax1.scatter(df.groupby('channel_name')['subscriberCount'].mean(), df.groupby('channel_name')['nbrMotInsulte'].sum())
ax1.set_title("Nombre d'insultes en fonction du nombre d'abonnés des chaînes", fontsize=20)
```

```python
slope2, intercept2, r_value, p_value, std_err = linregress(df.groupby('channel_name')['viewCount'].mean(),
                                                          df.groupby('channel_name')['nbrMotInsulte'].sum())
fitLine = slope2 * df.groupby('channel_name')['viewCount'].mean() + intercept2
fig2, ax2 = plt.subplots(1, 1, figsize=(20,6))
ax2.plot(df.groupby('channel_name')['viewCount'].mean(), fitLine, c='b')
ax2.scatter(df.groupby('channel_name')['viewCount'].mean(), df.groupby('channel_name')['nbrMotInsulte'].sum())
ax2.set_title("Nombre d'insultes en fonction du nombre de vues de la chaîne", fontsize = 20)
```

La tendance de ce scatter plot est ascendante, ce qui veut dire que globalement, plus il y a de vues sur une chaîne, plus il y a d'insultes, ce qui paraît logique.

```python
print('Nombre de vidéo par catégorie de chaîne', '\n', df.groupby('categorie_new').channel_name.count())
```

```python
# Tracé du nombre d'insultes en fonction de la ponctuation
# En laissant toutes les chaînes on se rend peu compte de la relation entre les deux. 
# On trace alors pour toutes les chaînes, séparément.
mpl.rcParams['figure.max_open_warning'] = 60
for name in df.channel_name.unique() :
    slope_excl, intercept_excl, r_value_e, p_value_e, std_err_e = linregress(df[df['channel_name'] == name]['nbrExclMark'],
                                                          df[df['channel_name'] == name]['nbrMotInsulte'])
    fitLine_excl = slope_excl * df[df['channel_name'] == name]['nbrExclMark'] + intercept_excl
    slope_quest, intercept_quest, r_value_q, p_value_q, std_err_q = linregress(df[df['channel_name'] == name]['nbrQuestMark'],
                                                          df[df['channel_name'] == name]['nbrMotInsulte'])
    fitLine_quest = slope_quest * df[df['channel_name'] == name]['nbrQuestMark'] + intercept_quest
    fig, ax = plt.subplots(1, 2, figsize=(20, 10))
    fig.suptitle(f'Insultes fonction de la ponctuation, pour {name} uniquement', fontsize=20)
    ax[0].scatter(df[df['channel_name'] == name]['nbrExclMark'], df[df['channel_name'] == name]['nbrMotInsulte'], color='r')
    ax[0].plot(df[df['channel_name'] == name]['nbrExclMark'], fitLine_excl, color='r')
    ax[0].set_title("Point d'exclamation", fontsize=15)
    ax[1].scatter(df[df['channel_name'] == name]['nbrQuestMark'], df[df['channel_name'] == name]['nbrMotInsulte'], color='g')
    ax[1].plot(df[df['channel_name'] == name]['nbrQuestMark'], fitLine_quest, color='g')
    ax[1].set_title("Point d'interrogation", fontsize=15)
```

Il est difficile de pouvoir tirer des informations intéressantes des graphiques ci-dessus : l y a trop de valeur entassées autour de l'origine pour réellement déceler une regréssion pertinente.

```python
# Proportion sum(insultes) / count(video) par categorie_new

prop = df[['nbrMotInsulte', 'categorie_new']].groupby('categorie_new').sum()
prop['nbr_video'] = df.groupby('categorie_new').video_id.count()
prop['Proportion'] = prop['nbrMotInsulte']/prop['nbr_video']
prop['Proportion'].plot(kind='bar')
```

Les vidéos de type 'Core' ont le taux d'insultes par vidéo le plus élevé. On peut penser que la 'catégorie_new' aura un certain poids dans l'élaboration de l'indice de toxicité.
